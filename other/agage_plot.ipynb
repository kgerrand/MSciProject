#### Code for plotting AGAGE station locations

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import seaborn as sns
```

```python
# Coordinates of the AGAGE monitoring stations (taken from https://agage.mit.edu/global-network)
# coordinates of sites explored
coordinates_explored = [(-9.9046, 53.3267),  # MHD
                        (126.1616, 33.2924)] # GSN
# coordinates of all sites
coordinates_sites = [(144.6894, -40.6833),   # CGO
                     (-59.4321, 13.1651),    # RPB
                     (7.985883, 46.547767),  # JFJ
                     (10.7014, 44.1932),     # CMN
                     (-124.141, 41.0541),    # THD
                     (11.8867, 78.9072)      # ZEP
                    ]


# creating world map
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(15,15))
sns.set(style='whitegrid')
sns.set_palette("colorblind")

# plotting coordinates
ax.scatter(*zip(*coordinates_explored), c='red', s=50, marker='o', zorder=2, label='Explored Sites')
ax.scatter(*zip(*coordinates_sites), c='blue', s=50, marker='o', zorder=2, label='Other Sites')

ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.set_global()
# ax.set_title('Map Showing the Locations of AGAGE Monitoring Stations', fontsize=15)
ax.legend(loc='upper left', fontsize=12)

plt.show()
```
