#### Code for plotting AGAGE station locations

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import seaborn as sns
```

```python
# Coordinates of the AGAGE monitoring stations (taken from https://agage.mit.edu/global-network)
# coordinates of sites explored
coordinates_explored = [(53.3267, -9.9046),  # MHD
                        (33.2924, 126.1616)] # GSN

labels_explored = ['MHD', 'GSN']

# coordinates of all sites
coordinates_sites = [(144.6894, -40.6833),   # CGO
                     (-59.4321, 13.1651),    # RPB
                     (7.985883, 46.547767),  # JFJ
                     (10.7014, 44.1932),     # CMN
                     (-124.141, 41.0541),    # THD
                     (11.8867, 78.9072)      # ZEP
                    ]

# creating world map
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(15,15))
sns.set(style='whitegrid')
sns.set_palette("colorblind")

# plotting explored sites and adding labels
for (lat, lon), label in zip(coordinates_explored, labels_explored):
    ax.scatter(lon, lat, color='red', s=50, marker='o', zorder=2)
    ax.annotate(label, (lon, lat), color='black', fontsize=15, ha='center', va='bottom', zorder=3,
                xytext=(-21, -9), textcoords='offset points', weight='bold')
    
# plotting other sites
ax.scatter(*zip(*coordinates_sites), c='blue', s=50, marker='o', zorder=2, label='Other Sites')

ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.set_global()
# ax.set_title('Map Showing the Locations of AGAGE Monitoring Stations', fontsize=15)

# creating legend
handles, labels = ax.get_legend_handles_labels()
handles = [handles[0], plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10)]
labels = ['Other Sites', 'Explored Sites']

ax.legend(handles=handles, labels=labels, loc='upper left', fontsize=14)

plt.show()
```
